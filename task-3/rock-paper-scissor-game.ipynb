```python
from tkinter import*
from PIL import Image,ImageTk
from random import randint

#main window
root =Tk()
root.title("Rock-Paper-Scissors")
root.configure(background="#9135D0")

#images
rock_img = ImageTk.PhotoImage(Image.open("rock2.png"))
paper_img = ImageTk.PhotoImage(Image.open("paper2.png"))
scissors_img = ImageTk.PhotoImage(Image.open("scissors2.png"))

#show image
user_img = Label(root,image=rock_img,bg="#9135D0")
comp_img = Label(root,image=scissors_img,bg="#9135D0")
user_img.grid(row=1,column=0)       
comp_img.grid(row=1,column=4) 

#score
userscore = Label(root,text=0,font=("Arial",40),bg="#9135D0",fg="white")
compscore = Label(root,text=0,font=("Arial",40),bg="#9135D0",fg="white")
userscore.grid(row=1,column=1)
compscore.grid(row=1,column=3)

#Labels
user = Label(root,font=("Arial",24,"bold"),text="USER",bg="#9135D0",fg="white").grid(row=0,column=1)
comp = Label(root,font=("Arial",24,"bold"),text="COMPUTER",bg="#9135D0",fg="white").grid(row=0,column=3)

#message
msg = Label(root,font=("Arial",22),bg="#9135D0",fg="white")
msg.grid(row=4,column=2)

#update message
def updateMessage(x):
   msg['text'] = x
   
#update userscore
def updateUserScore():
    score = int(userscore["text"]) 
    score += 1
    userscore["text"] = str(score)
    
#update compscore
def updateCompScore():
    score = int(compscore["text"]) 
    score += 1
    compscore["text"] = str(score)        
   
#check winner
def checkWinner(user,computer):
    if user == computer:
        updateMessage("It's a Tie!!")
    elif user == "rock":
        if computer == "scissors":
            updateMessage("You win!")
            updateUserScore()
        else:
            updateMessage("You loose!!") 
            updateCompScore()
    elif user == "paper":
        if computer == "rock":
            updateMessage("You win!")  
            updateUserScore()
        else:
            updateMessage("You loose!!") 
            updateCompScore()  
    elif user == "scissors":
        if computer == "paper":
            updateMessage("You win!")
            updateUserScore()
        else:
            updateMessage("You loose!!") 
            updateCompScore()  
    else:
        pass                            
   
#choices
choices = ["rock","paper","scissors"]
def updateChoice(x):
    
    #for computer
    compChoice = choices[randint(0,2)]
    if compChoice == "rock":
        comp_img.configure(image=rock_img)
    elif compChoice=="paper":
        comp_img.configure(image=paper_img)
    else:
        comp_img.configure(image=scissors_img)    
        
    #for user
    if x=="rock":
        user_img.configure(image=rock_img)
    elif x=="paper":
        user_img.configure(image=paper_img)
    else:
        user_img.configure(image=scissors_img)  
        
    checkWinner(x,compChoice)    
            

#choices button
rock = Button(root,width=20,height=3,text="ROCK",font=("Arial",13,"bold"),bg="#FF3E4D",fg="white",command = lambda:updateChoice("rock")).grid(row=2,column=1)
paper = Button(root,width=20,height=3,text="PAPER",font=("Arial",13,"bold"),bg="#FAD02E",fg="white",command = lambda:updateChoice("paper")).grid(row=2,column=2)
scissors = Button(root,width=20,height=3,text="SCISSORS",font=("Arial",13,"bold"),bg="#0ABDE3",fg="white",command = lambda:updateChoice("scissors")).grid(row=2,column=3)

root.mainloop()
```
